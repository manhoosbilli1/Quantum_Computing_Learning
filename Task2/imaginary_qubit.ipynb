```python
from qiskit import QuantumCircuit 
from qiskit.quantum_info import Statevector 
from qiskit.visualization import plot_bloch_multivector, plot_bloch_vector 

# Illustrating the problem with drawing a statevector that contains an imaginary value. 

qc = QuantumCircuit(1) 
qc.x(0) # to flip the bit to 1. 
qc.y(0) # to apply y gate to state 1, flips it and adds a phase and minus size.  = -i|0> 
state = Statevector(qc) # create a statevector from given circuit 
display(state.draw('latex', prefix="output = "))
display(state.draw('bloch')) 
```
